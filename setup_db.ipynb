```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import config
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#Extract CSV into DataFrame
tvshows_df = pd.read_csv("US_Data/model_data.csv")
tvshows_df
```

```python
#Create Database connection
connection_string = f"postgres:{config.password}@localhost:5432/tvtimedb"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Use pandas to Load  csv converted DataFrames into database
tvshows_df.to_sql(name='tvshows', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from tvshows', con=engine).head()
```
