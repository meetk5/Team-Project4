```python
def predict(prediction):
    if (prediction < 2000):
        return_string = "Sorry! The show may not do well with the audience!"
    elif (prediction < 10000):
        return_string = "Your show will get an average commercial success!"
    elif (prediction < 100000):
        return_string = "Congratulations! Your show will be a GREAT SUCCESS!"
    elif (prediction < 500000):        
        return_string = "CONGRATULATIONS! You show will be a BLOCKBUSTER!"
    return return_string
```
