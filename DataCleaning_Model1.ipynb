```python
allepisodedf = pd.read_csv("Resources/all_episodes.csv")
allepisodedf.head()
```

```python
allepisodedf.shape
```

```python
allepisodedf = allepisodedf.drop(["episode_id","episode","title","description","rating","times_watched"],axis = 1)
allepisodedf.head()
```

```python
modeldf = tvshowdf1.merge (grouped_episodedf, on = "show_id")
modeldf.head()
```
